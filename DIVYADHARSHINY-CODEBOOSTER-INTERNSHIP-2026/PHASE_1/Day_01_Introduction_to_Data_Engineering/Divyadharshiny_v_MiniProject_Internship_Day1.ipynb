```python
print("Total number of students :", len(df))
print("Total number of columns :", len(df.columns))
print("Unique departments :", df['department'].nunique())
```

```python
dept_count = df["department"].value_counts()
print(dept_count)
```

```python
print("Highest Math Score :", df["math_score"].max())
```

```python
print("Lowest Math Score :", df["math_score"].min())
```

```python
print("Average Math Score :", df["math_score"].mean())
```

```python
print("Average Programming Score :", df["programming_score"].mean())
```

```python
print("Average Attendance :", df["attendance_percentage"].mean())
```

```python
df["total_score"] = df["math_score"] + df["programming_score"]
```

```python
gender_avg = df.groupby("gender")["total_score"].mean()
print(gender_avg)
```

```python
top_students = df.sort_values(by="total_score", ascending=False).head(5)
print(top_students[["name", "department", "total_score"]])
```

```python
low_attendance = df[df["attendance_percentage"] < 75]
if len(low_attendance) > 0:
    print(low_attendance[["name", "department", "attendance_percentage"]])
else:
    print("All students have attendance above 75%")
```

```python
# PRACTICE QUESTION
electronics_count = len(df[df["department"] == "Electronics"])
print("Students from Electronics Department :", electronics_count)
```

```python
female_attendance = df[df["gender"] == "Female"]["attendance_percentage"].mean()
print("Average attendance of female students :", female_attendance)
```

```python
lowest_prog = df.loc[df["programming_score"].idxmin()]
print("Student Name :", lowest_prog["name"])
print("Programming Score :", lowest_prog["programming_score"])
```

```python
above_80 = len(df[df["math_score"] > 80])
print("Students scored above 80 in math :", above_80)
```

```python
avg_total = df[df["attendance_percentage"] > 90]["total_score"].mean()
print("Average total score :", avg_total)
```
